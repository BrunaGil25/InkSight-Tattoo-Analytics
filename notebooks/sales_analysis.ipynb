```python
#  Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Load sales data
df = pd.read_csv('../data/sales_mock.csv')

#  Setup
sns.set(style='whitegrid')
df['date'] = pd.to_datetime(df['date'])

#  Total sales per product
product_counts = df['product'].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=product_counts.index, y=product_counts.values, palette='Set2')
plt.title('Total Sales per Product')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()

#  Sales by year and product
year_product = df.groupby(['year', 'product']).size().unstack().fillna(0)
year_product.plot(kind='bar', stacked=True, colormap='Set2')
plt.title('Sales Volume by Year and Product')
plt.xlabel('Year')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()

#  Gender distribution per product
gender_product = df.groupby(['product', 'gender']).size().unstack().fillna(0)
gender_product.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Gender Distribution by Product')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()

#  Price distribution per product
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='product', y='price', palette='Set2')
plt.title('Price Distribution by Product')
plt.xlabel('Product')
plt.ylabel('Price ()')
plt.tight_layout()
plt.show()
```
